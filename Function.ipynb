```python
def count(lst):
    even=0
    odd=0
    for i in lst:
        if i%2 ==0:
            even+=1
        else:
            odd+=1
    return even,odd
lst=[1,2,3,4,5,6,7,8,9]
even,odd = count(lst)
print('even numbers:',even)
print('odd numbers',odd)
```

```python
def is_even(n):
    return n % 2== 0
nums=[3,45,5,8,2,6,9,4,6]
evens= list(filter(is_even, nums))
print(evens)
```

```python
from functools import reduce
def add_all(a,b):
    return a+b
nums=[3,4,5,6,7,8,9]
evens=list(filter(is_even, nums))
double= list(map(lambda n: n*2,evens))
sums=reduce(add_all,double)
print(sums)
    
```
