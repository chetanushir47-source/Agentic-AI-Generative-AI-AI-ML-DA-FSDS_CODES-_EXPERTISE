```python
import tensorflow as tf
from tensorflow.keras.preprocessing_imaget import ImageDataGenerator
from tensorflow.keras.preprocessing_imaget import load_img,img_to_array
```

```python
datagen= ImageDataGenerator(
    rotation_range =40,
    width_shift_range=0.4,
    height_shift_range=0.4,
    shear_range= 0.4,
    zoom_range= 0.4,
    horizantal_flip = True,
    rescale= None,
    fill_mode= 'wrap'   # reflect, nearest ,constant,wrap
)
```

```python
img= load_img(r"C:\Users\hp\Desktop\training\sad\sad2.jpg")
img
```

```python
x= img_to_array(img)
x=x.reshape((1,)+x.shape)

i=0
for batch in datagen.flow(x,batch_size= 1,
                          save_to_dir=r""
                          save_prefix='dog',ave_format='jpeg'):
    i+=1
    if i>30:
        break
```
